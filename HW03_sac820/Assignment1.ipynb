### Import Statements


```python
from __future__ import print_function
__author__= 'sac820'

import numpy as np
import pylab as pl
import seaborn
    
%pylab inline
```

### List the different kinds of distributions

1. Chi Squared
2. Gaussian 
3. Poisson 
4. Binomial
5. Gamma 


### Seed the pseudrandom number generator to allow reproducibility

```python
np.random.seed(50)
```

### Set a mean for all distributions

```python
mean = 100

##For the Binomial, you are looking not at the mean but at the probability of success over a number of trials
probabilitySuccess = .5
numberTrials = 20

## For the Poisson, you are lookign at the expected number of occurrences lambda
lbda = mean

##For Chi Square Distribution, you use degrees of freedom
degreesFreedom = mean
```

### Create an array of random integers to set different sample sizes for the distributions

```python
#This method of creating a set of random sample size numbers has a broader range of values which I believe
#is better suited for showing the CLT and Law of Large Numbers
sampleSize = (2000 / (np.array(range(1, 100)))).astype(int)
print (sampleSize, sampleSize.shape)
```

### Create empty nested dictionary to host distributions

```python
distributionDictionary = {}
```

# 1. CHI SQUARED

```python
distributionDictionary['chisq'] = {} 

for n in sampleSize:
    distributionDictionary['chisq'][n] = np.random.chisquare(degreesFreedom, size = n)
    
distributionDictionary['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in distributionDictionary['chisq']:
    if not type(nn) == str:
        distributionDictionary['chisq']['means'][nn] = distributionDictionary['chisq'][nn].mean()
        axchisq_mu_n.plot(nn, distributionDictionary['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(sampleSize), max(sampleSize)], [degreesFreedom, degreesFreedom], 'k')
```

##### Figure 1: Plotting the sample size against the sample mean in comparison to the actual mean shows that as your sample size increases, the closer your sample mean gets to the population mean. This is the Law of Large Numbers.

```python
#Create a list of the means
allmeans = list(distributionDictionary['chisq']['means'].values())

#Plot the histogram of samples
pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30, normed=0)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

##### Figure 2: Creating a histogram of the sample means collected shows the Central Limit Theorem: the distribution of the sample means is Gaussian.

```python
#Fit a normal curve to the histogram

#I know this is not very Pythonic but I was having some trouble getting Pylab and this to get along. 
#There is another method using Scipy.curve but it's way more complicated than this.
from scipy.stats import norm

#Plot the probability density function
pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30, normed=1)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)

# Fit a normal distribution to the data
mu, std = norm.fit(allmeans)

# Plot the PDF.
xmin, xmax = pl.xlim()
x = np.linspace(xmin, xmax)
p = norm.pdf(x, mu, std)
pl.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
pl.title(title)
```

##### Figure 3: Plotting the probability density function and then fitting a Gaussian curve over the distribution gives us a reasonable idea that the Central Limit Theorem shows that the distribution of sample means of a population is normal.

# 2. Gaussian


```python
distributionDictionary['gaus'] = {} 

for n in sampleSize:
    distributionDictionary['gaus'][n] = np.random.normal(mean, size = n)
    
distributionDictionary['gaus']['means'] = {}
axgauss_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in distributionDictionary['gaus']:
    if not type(nn) == str:
        distributionDictionary['gaus']['means'][nn] = distributionDictionary['gaus'][nn].mean()
        axgauss_mu_n.plot(nn, distributionDictionary['gaus']['means'][nn], 'o')
        axgauss_mu_n.set_xlabel('sample size', fontsize=18)
        axgauss_mu_n.set_ylabel('sample mean', fontsize=18)
        axgauss_mu_n.set_title('Gaussian', fontsize=18)
        axgauss_mu_n.plot([min(sampleSize), max(sampleSize)], [mean, mean], 'k')
```

##### Figure 4: This shows the Law of Large Numbers which means as the sample size gets larger, the mean of your sample means gets closer to the population mean.

```python
allmeans = list(distributionDictionary['gaus']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

##### Figure 5: This histogram shows that the distribution of sample means is normal. 

```python
#Fit a normal curve to the histogram

#I know this is not very Pythonic but I was having some trouble getting Pylab and this to get along. 
#There is another method using Scipy.curve but it's way more complicated than this.
from scipy.stats import norm

#Plot the probability density function
pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30, normed=1)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)

# Fit a normal distribution to the data
mu, std = norm.fit(allmeans)

# Plot the PDF.
xmin, xmax = pl.xlim()
x = np.linspace(xmin, xmax)
p = norm.pdf(x, mu, std)
pl.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
pl.title(title)
```

##### Figure 6: Just like above, fitting a Normal curve to it demonstrates thatthe distribution of the sample mean is somewhat bell shaped

# 4. Binomial

```python
distributionDictionary['binomial'] = {} 

for n in sampleSize:
    distributionDictionary['binomial'][n] = np.random.binomial(numberTrials, probabilitySuccess, size = n)
    
distributionDictionary['binomial']['means'] = {}
axbi_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in distributionDictionary['binomial']:
    if not type(nn) == str:
        distributionDictionary['binomial']['means'][nn] = distributionDictionary['binomial'][nn].mean()
        axbi_mu_n.plot(nn, distributionDictionary['binomial']['means'][nn], 'o')
        axbi_mu_n.set_xlabel('sample size', fontsize=18)
        axbi_mu_n.set_ylabel('sample mean', fontsize=18)
        axbi_mu_n.set_title('Binomial', fontsize=18)
        axbi_mu_n.plot([min(sampleSize), max(sampleSize)], [numberTrials*probabilitySuccess, numberTrials*probabilitySuccess], 'k')
```

##### Figure 10: A binomial distribution is the number of successes over a number of trials. So a 50% probability of success over 20 trials will get you 10 trials. Think kof flipping a coin. This shows the Law of Large Numbers. If you flip it only a few times, you're going to get a spread of values (so many heads, so many tails). But if you do it enough times, you get the 50/50. This is actually part of what drives the frequentist view of statistics. Sorry, Bayes.
