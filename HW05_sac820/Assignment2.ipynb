## AD and KS test

**Test the following:**

1. A normal distribution N(0,1) against a Gaussian model, 
2. A Poisson distribution with small $\lambda$ or a binomial _with small np_  against a Gaussian model. 


**Expected outcomes:** In the first case the Gaussian is indeed Gaussian and in the second case that the binomial with small np is _not_ Gaussian.

### Normal against Normal using AD and KS

#### Generate a normal distribution

```python
alpha = 0.05
np.random.seed(999)
dist_n = np.random.randn(1000) 
```

#### Calculate and print the test results

```python
print ("Normal on normal KS test: ", scipy.stats.kstest(dist_n,'norm'))
print ("Normal on normal AD test: ", scipy.stats.anderson(dist_n, dist='norm'))
```

#### Analyze the test results

**KS Test:** The p-value is well above 0.05 meaning we cannot reject the $H_0$ that the sample comes from a Gaussian distribution.

**AD Test:** With an $\alpha$ of 0.05, the critical value will be 0.784. With the statistic at 0.26, we cannot reject the $H_0$ that the sample comes from a Gaussian distribution.

### Poisson against Normal using AD and KS

#### Generate a Poisson distribution with high $\lambda$ 

```python
np.random.seed(999)
```

```python
#poisson with 1000 elements, lambda=100 (high)
dist_p = np.random.poisson(100, 1000)
```

#### Plot a Poisson against a Gaussian

```python
pl.figure()
pl.hist(dist_p, alpha=0.5, label="Poisson")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("Similar distributions in shape\nbut different means and stretch")
pl.show()
```

#### Calculate and print test results

```python
print ("Poisson on normal KS test: ", scipy.stats.kstest(dist_p,'norm'))
print ("Poisson on normal AD test: ", scipy.stats.anderson(dist_p, dist='norm'))
```

In this case, the means are different, so the distributions will be different for the KS and AD tests. We want to test a Poisson distribution with a high mean shape because this appears more Gaussian. 

We want this Poisson distribution to have the same mean as the Gaussian: $\mu=0$. 

Similarly, we want the standard deviation to be 1 instead of $\sqrt{\lambda} = \sqrt{100}.$

#### Fix the mean and standard deviation of the Poisson distribution

```python
dist_p = (np.random.poisson(100, 1000) - 100) * 0.1 #-100 fixes mean, *0.1 fixes stdev
```

#### Plot Poisson against Gaussian moving the Poissonian mean 

```python
pl.figure()
pl.hist(dist_p, alpha=0.5, label="Poisson")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("Similar distributions")
pl.show()
```

#### Calculate and print the test results

```python
print ("Poisson on normal KS test: ", scipy.stats.kstest(dist_p,'norm'))
print ("poisson on normal AD test: ", scipy.stats.anderson(dist_p, dist='norm'))
```

**KS Test:** No longer rejecting the $H_0$ at $\alpha=0.05$ with a p-value of around 0.015

```python
threshold = scipy.stats.anderson(dist_n, 
                                 dist='norm')[1][scipy.stats.anderson(dist_n, 
                                                                      dist='norm')[2]==[5.0]]
print ("AD test threshold for p = 0.05", threshold[0])
```

**AD Test:** Statistics AD=0.93 larger than the threshold for $\alpha=0.05$

### Binomial against Normal using AD and KS

#### Generate a binomial distribution

```python
#binomial with n=10, p=0.5 => mean=np=5, stdev=sqrt(np(1-p))
n, p = 200, 0.5
dist_b = (np.random.binomial(n, p, 1000) - n * p) / np.sqrt(n * p * (1 - p))  #subtract mean divide by stdev
```

#### Plot a Binomial against a Gaussian

```python
pl.figure()
pl.hist(dist_b, alpha=0.5, label="Binomial")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("Similar distributions")
pl.show()
```

#### Calculate and print test results

```python
print ("Binomial on normal KS test: ", scipy.stats.kstest(dist_b,'norm'))
print ("Binomial on normal AD test: ", scipy.stats.anderson(dist_b, dist='norm'))
```

#### Analyze test results

**N.B.**: the statistics value will be **larger** if the distributions are **not** likely to be related. 




<center>**Null Hypotheses according to manuals**<center>

KS Test: Under the null hypothesis the two distributions are identical such that $G(x)=F(x)$

AD Test: Sample is drawn from a population that follows a particular distribution.

<center>**Rewritten null hypotheses**<center>

KS Test: The distributions are identical

AD Test: The sample distribution is drawn from a population which follows the specificed distribution.

**KS Test:** The p-value is well below 0.05 allowing us to reject the null hypothesis that the distributions are identical.

**AD Test:** The statistic is well above the critical value set by the threshold of 0.05 allowing us to reject the null hypothesis that the sample distribution is drawn from the Normal distribution.

## KL Divergence Test

<center>__ Empirical Distribution__<center> 

The test wants to know the distribution _y_ value at a location _x_.  We have to pass it the values of the _distribution_ at some x's. It does not need the histogram though since we need a normalized histogram of the distribution. In other words, we need to derive a probability distribution function (PDF) from the data. 

<center>__Model__<center>

Then we have to pass the test a function for the comparison distribution evaluated at the same _x_-values where the empirical distribution is evaluated (not just the name of a distribution like for KS and AD).

**NOTE on KL divergence**: the KL divergence is a measure of how much approximation you have to do to represent one distribution with the other in terms of information content. You can see what it looks point by point, before you integrate and get a single distance number.

### Create PDFs for the three different distributions

```python
# np.histogram returns the value of the histogram (y) and the edges of the bins (x). 
# Notice: the edges, not the centers
distpdf_n, mybins_n, = np.histogram(dist_n, density=True)
distpdf_b, mybins_b, = np.histogram(dist_b, density=True)
distpdf_p, mybins_p, = np.histogram(dist_p, density=True)

#Notice: the extra comma on the left side of the '=' sign: 
#that tells numpy take the first two values returned, 
#and throw away the rest
```

Then we have to pass it the functional form of a normal PDF, which is returned by the function scipy.stats.norm for the example evaluated at the bin centers of the histogram from before.

```python
#get the bin centers
bincenters_n = mybins_n[:-1] + 0.5*(mybins_n[1] - mybins_n[0])
bincenters_b = mybins_b[:-1] + 0.5*(mybins_b[1] - mybins_b[0])
bincenters_p = mybins_p[:-1] + 0.5*(mybins_p[1] - mybins_p[0])
```

### Test all three distributions with KL

The KL test returns the "entropy". To figure out what the number means though in terms of NULL rejection, help yourself with the fact that you know what the result should be for these three distributions!


```python
print ("Normal on normal for KL test: ", 
    scipy.stats.entropy(distpdf_n, 
                            scipy.stats.norm.pdf(bincenters_n)) )
print ("Poisson on normal for KL test: ", 
       scipy.stats.entropy(distpdf_p, 
                                scipy.stats.norm.pdf(bincenters_p)))

print ("Binomial on normal for KL test: ", 
       scipy.stats.entropy(distpdf_b, 
                                scipy.stats.norm.pdf(bincenters_b)))

#you can interpret this as a distance: 
# it increases as the distributions diverge
```

<center>**To reject or not to reject**<center>

In the KS and AD tests, we did not reject the nulls for the Gaussian and Poisson distributions but did reject the null hypothesis for the binomial. Thus, it seems, as the entropy increases, we can think of the p-value decreasing. We know then, at the least, that entropy values from .17 and above allow us to reject the null while values of .01 and below do not. 

## $\chi^2$ Test

To use the scipy.stats.chisquare test, you must pass it an array of the observed frequencies and then the expected frequencies. The chi square test tests the null hypothesis that the categorical data has the given frequencies. This does not work very well as this test is intended for categorical data.

```python
print(scipy.stats.chisquare(distpdf_n, scipy.stats.norm.pdf(bincenters_n)))
print(scipy.stats.chisquare(distpdf_p, scipy.stats.norm.pdf(bincenters_p)))
print(scipy.stats.chisquare(distpdf_b, scipy.stats.norm.pdf(bincenters_b)))
```

## Increasing the product of $n$ and $p$ in a binomial distribution

With the binomial distribution, we can change $n*p$. A larger $n*p$ will make the distribution look more "normal" (i.e. Gaussian). In a loop test, we see that the binomial tends to a normal distribution as you increase the value of $n*p$.

### Loop over different values of n with each test

```python
narray = list(range(1,501))

#list for KS test statistics and pvalues
ks_s_b = list(range(500))
ks_p_b = list(range(500))

#list for AD test statistic
ad_b = list(range(500))

#list for KL entropy values
kl_b = list(range(500))

#list for Chisquare statistics and p-values
chi_s = list(range(500))
chi_p = list(range(500))
```

```python
for i,n in enumerate(narray):
    #parameter for the binomial, my arbitrary choice
    p = 0.5 
    
    #generate the distribution
    dist = (np.random.binomial(n, p, 1000) - n * p ) / np.sqrt(n * p * (1-p))
    
    #run the tests. 
    ks_s_b[i] = scipy.stats.kstest(dist,'norm')[0]
    ks_p_b[i] = scipy.stats.kstest(dist,'norm')[1]
    ad_b[i] = scipy.stats.anderson(dist, dist='norm')[0]
  
    #For KL and Pearson's chisq I have to simulate the normal distribution as well
    mybins = np.linspace(min(dist), max(dist), 10) 
    bincenters = mybins[:-1] + 0.5 * (mybins[1] - mybins[0])

    kl_b[i] =  scipy.stats.entropy(np.histogram(dist, bins=mybins)[0] / 1000., 
                                    scipy.stats.norm.pdf(bincenters))
    
    chi_s[i] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0] / 1000., 
                                    scipy.stats.norm.pdf(bincenters) + n)[0]
    chi_p[i] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0] / 1000., 
                                    scipy.stats.norm.pdf(bincenters) + n)[1]
```

### Plot the different tests with increasing values

```python
fig = pl.figure(figsize = (15,10))
fig.suptitle('Figure 1: Binomial Distribution Goodness of Fit to Normal Distribution as N Increases')
plt.xlabel('Size of N')
plt.ylabel('Test Statistic Value')

fig.add_subplot(221)
pl.plot(narray, ks_s_b, label='KS statistics')
pl.plot(narray, ks_p_b, label='KS p-value')
pl.title('Figure 1A: KS Test')
pl.legend()

fig.add_subplot(222)
pl.plot(narray, ad_b,  label='AD statistics')
pl.plot([narray[0], narray[-1]],[threshold, threshold], label="threshold")
pl.ylim(0,10) #limit the y range or you cannot see the relevant part
pl.title('Figure 1B: AD Test')
pl.legend()

fig.add_subplot(223)
pl.plot(narray, kl_b, label='K-L (entropy)')
pl.ylim(0,0.2)
pl.title('Figure 1C: KL Test')
pl.legend();

fig.add_subplot(224)
pl.title('Figure 1D: Chi Squared Test')
pl.plot(narray, chi_s, label='Chi Squared Statistic')
pl.plot(narray, chi_p, label='Chi Squared p-value')
pl.ylim(0,5000)
pl.legend();
```


**Figure 1: Three plots of KS, AD, and KL statistics plotted against an increasing n of a binomial distribution. It shows that as n increases, it becomes more Gaussian**

**Figure 1A:** As the product of n and p increases, we see that the KS statistic decreases rapidly at first and then slowly. At the same time, the p-value increases almost linearly. From previous tests we know that for the KS test, a low KS statistic and a high p-value shows that the two distributions are identical. So, as the product of n and p increases, we see that the binomial distribution becomes more and more normal.

**Figure 1B:** As the product of n and p increases, we see that the AD statistic decreases rapidly at first and then slowly. Plotted against it is the threshold for significance at an $alpha$ of 0.05. Thus, we can see that as the product of n and p increases, the binomial distribution gets closer and closer to the threshold and surpasses it becoming more normal.

**Figure 1C:** As the product of n and p increases, we see that the entropy calculated in the KL test drops off rapidly and then slowly again showing that as this product increases, the informational difference between the two distributions becomes smaller and smaller with the limit approaching zero. 


## Increasing the $\lambda$ parameter in a Poisson distribution

### Loop over different values of $\lambda$

```python
narray = list(range(1,501))

ks_s_p = list(range(500))
ks_p_p = list(range(500))

ad_p = list(range(500))

kl_p = list(range(500))

chi_ps = list(range(500))
chi_pp = list(range(500))
```

```python
n = 10
for i,n in enumerate(narray):
    p = 0.5 #parameter for the binomial, my arbitrary choice
    #generate the distribution
    dist = (np.random.poisson(lam=n, size=100) - n) / np.sqrt(n)
    
    #dist = np.random.poisson(lam=n, size=100)
    ks_s_p[i] = scipy.stats.kstest(dist,'norm')[0]
    ks_p_p[i] = scipy.stats.kstest(dist,'norm')[1]
    ad_p[i] = scipy.stats.anderson(dist, dist='norm')[0]

    mybins = np.linspace(min(dist), max(dist), 10) 
    bincenters = mybins[:-1] + 0.5 * (mybins[1] - mybins[0])

    kl_p [i] =  scipy.stats.entropy(np.histogram(dist, bins=mybins)[0] / 1000, 
                                    scipy.stats.norm.pdf(bincenters))
    chi_ps[i] =  scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0], 
                                    scipy.stats.norm.pdf(bincenters) + n)[0]
    chi_pp[i] =  scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0], 
                                    scipy.stats.norm.pdf(bincenters) + n)[1]
```

### Plot the different tests with increasing values

```python
fig = pl.figure(figsize = (15,10))
fig.suptitle('Figure 2: Poisson Distribution Goodness of Fit to Normal Distribution as Lambda Increases')
plt.xlabel('Size of Lambda')
plt.ylabel('Test Statistic Value')

fig.add_subplot(221)
pl.plot(narray, ks_s_p, label='KS statistics')
pl.plot(narray, ks_p_p, label='KS p-value')
pl.title('Figure 2A: KS Test')
pl.legend()

fig.add_subplot(222)
pl.plot(narray, ad_p,  label='AD statistics')
pl.plot([narray[0], narray[-1]],[threshold, threshold], label="threshold")
pl.ylim(0,10) #limit the y range or you cannot see the relevant part
pl.title('Figure 2B: AD Test')
pl.legend()

fig.add_subplot(223)
pl.title('Figure 2C: KL Test')
pl.plot(narray, kl_p, label='K-L (entropy)')
pl.legend();

fig.add_subplot(224)
pl.title('Figure 2D: Chi Squared Test')
pl.plot(narray, chi_ps, label='Chi Statistic')
pl.plot(narray, chi_pp, label='Chi p-value')
pl.ylim(0,5000)
pl.legend();
```


**Figure 2: Three plots of KS, AD, and KL statistics plotted against an increasing $\lambda$ of a Poisson distribution. It shows that as $\lambda$ increases, it becomes more Gaussian**

From previous tests we know that for the KS test, a low KS statistic and a high p-value shows that the two distributions are identical. So, as $\lambda$ increases, we see that the Poisson distribution becomes more and more normal.

**Figure 2B:** As $\lambda$ increases, we see that the AD statistic decreases rapidly at first and then slowly. Plotted against it is the threshold for significance at an $alpha$ of 0.05. Thus, we can see that as $\lambda$ increases, the Poisson distribution gets closer and closer to the threshold and surpasses it becoming more normal.

**Figure 2C:** As $\lambda$ increases, we see that the entropy calculated in the KL test drops off rapidly and then slowly again showing that as this product increases, the informational difference between the two distributions becomes smaller and smaller with the limit approaching zero.
